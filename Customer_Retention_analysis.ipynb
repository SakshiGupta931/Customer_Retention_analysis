```python
import pandas as pd
```

```python
df = pd.read_csv("customer_retention_dummy_data.csv")
```

**EDA**

```python
# Retention rate comparison
retention_rates = df.groupby('Loyalty_Member')['Retained'].mean()
print(retention_rates)
```

**Visualisation**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x='Loyalty_Member', y='Retained', data=df)
plt.title("Retention Rate by Loyalty Membership")
plt.xticks([0, 1], ['Non-Member', 'Member'])
plt.ylabel("Retention Rate")
plt.show()
```

Hypothesis Testing

```python
from statsmodels.stats.proportion import proportions_ztest
import numpy as np

```

```python
# Count of retained customers
count = np.array([
    df[df['Loyalty_Member'] == 1]['Retained'].sum(),
    df[df['Loyalty_Member'] == 0]['Retained'].sum()
])

# Total number of customers in each group
nobs = np.array([
    df[df['Loyalty_Member'] == 1].shape[0],
    df[df['Loyalty_Member'] == 0].shape[0]
])

# Perform one-tailed Z-test
stat, pval = proportions_ztest(count, nobs, alternative='larger')
print(f"Z-statistic: {stat:.3f}")
print(f"P-value: {pval:.4f}")
```
